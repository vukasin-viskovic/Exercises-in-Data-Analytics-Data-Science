##### Object-oriented API method

- The more flexible of the two methods

```python
fig = plt.figure() # figure-method

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, heigth
                                          # in relation to the blank canvas!

axes.plot(x, y)
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, heigth
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # left, bottom, width, heigth

axes1.plot(x, y, 'r-')
axes1.set_title('Larger plot')

axes2.plot(y, x)
axes2.set_title('Smaller plot')
```

##### Line color

##### `linewidth`

- Shorthand: `lw`

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, color = '#FF8C00', linewidth = 1) # default line width
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 2) # ... and indexed

axes[0,0].plot(x, y, color = '#FF1C22', lw = 1)
axes[0,1].plot(x, y, color = '#FF0C00', lw = 2) # twice as thick

axes[1,0].plot(x, y, color = '#FF3C99', lw = 10)
axes[1,1].plot(x, y, color = '#FF8C00', lw = 0.5) # two times thinner
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 4) # ... and indexed

axes[0].plot(x, y, color = 'orange', linewidth = 3, alpha = 1)
axes[1].plot(x, y, color = 'orange', linewidth = 3, alpha = 0.7)
axes[2].plot(x, y, color = 'orange', linewidth = 3, alpha = 0.5)
axes[3].plot(x, y, color = 'orange', linewidth = 3, alpha = 0.1)

plt.tight_layout()
```

##### `linestyle`

- Shorthand: `ls`

```python
fig, axes = plt.subplots(nrows = 1, ncols = 4) # ... and indexed

axes[0].plot(x, y, color = 'orange', lw = 3, linestyle = '--') # default: ls = '-'
axes[1].plot(x, y, color = 'orange', lw = 3, linestyle = '-.')
axes[2].plot(x, y, color = 'orange', lw = 3, linestyle = ':')
axes[3].plot(x, y, color = 'orange', lw = 3, linestyle = 'steps')

plt.tight_layout()
```

**MatplotlibDeprecationWarning:** Passing the drawstyle with the linestyle as a single string is deprecated since Matplotlib 3.1 and support will be removed in 3.3; please pass the drawstyle separately using the drawstyle keyword argument to Line2D or set_drawstyle() method (or ds/set_ds()).

https://matplotlib.org/stable/api/_as_gen/matplotlib.lines.Line2D.html

- drawstyle or ds: {'default', 'steps', 'steps-pre', 'steps-mid', 'steps-post'}
- default: 'default'

```python
fig, axes = plt.subplots(nrows = 1, ncols = 4) # ... and indexed

axes[0].plot(x, y, color = 'orange', lw = 3, ds = 'default')
axes[1].plot(x, y, color = 'orange', lw = 3, ls = ':')
axes[2].plot(x, y, color = 'orange', lw = 3, ls = '--')
axes[3].plot(x, y, color = 'orange', lw = 3, ds = 'steps', ls = '-.')

plt.tight_layout()
```
